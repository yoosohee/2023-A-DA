```python
import matplotlib.pyplot as plt
import numpy as np

N=8
y=np.zeros(N)

x1=np.linspace(0,10,N, endpoint=True) #10  8  ,  7
x2=np.linspace(0,10,N, endpoint=False) #10  8  ,  8
plt.plot(x1,y+0.5,'o',label='endpoint=True')
plt.plot(x2,y,'.',label='endpoint=False')

plt.legend()
plt.ylim([-0.5,1]) #Y  
plt.show()
```
