```python
import csv

f=open('subwayfee.csv', encoding='cp949')
data=csv.reader(f)

header=next(data)
print(header)

rate=0
max_rate=0

for row in data:
    for i in range(4,8):
        row[i]=int(row[i].replace(',',""))
    
    if row[6]!=0 and (row[4]+row[6]>100000):
        rate=row[4]/(row[4]+row[6])
        if rate>0.92:
            max_rate=rate
            print(row, round(max_rate,2))

f.close()
```

```python
import csv
f=open('subwayfee.csv', encoding='cp949')
data=csv.reader(f)

header=next(data)

max_people=[0]*4
max_station=['']*4
label=['(유임승차)','(유임하차)','(무임승차)','(무임하차)']

for row in data:
    for i in range(4,8):
        row[i]=int(row[i].replace(',',""))
        if row[i]>max_people[i-4]:
            max_people[i-4]=row[i]
            max_station[i-4]=row[3]+''+row[1]

f.close()

for i in range(4):
    print(label[i]+''+max_station[i]+':',max_people[i],'명')
```

```python
import csv
import matplotlib.pyplot as plt

f=open('subwaytime.csv', encoding='cp949')
data=csv.reader(f)

header=next(data)
next(data)

s_in=[0]*24
s_out=[0]*24

for row in data:
    for i in range(4,52):
        row[i]=row[i].replace(',',"")

    row[4:]=map(int, row[4:])
    for i in range(24):
        s_in[i]=s_in[i] +row[2*i+4] 
        s_out[i]=s_out[i]+row[2*i+5]

f.close()     

plt.rc('font', family='Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label='승차')
plt.plot(s_out, label='하차')
plt.legend()
plt.grid(True)
plt.xticks(range(24), range(4,28))
plt.show()
```
